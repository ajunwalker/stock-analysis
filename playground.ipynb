```python
def print_xml_nodes(node, depth=0):
	"""Recursively prints XML nodes with hierarchical indentation"""
	indent = '│   ' * depth + ('└── ' if depth > 0 else '')
	# Display node metadata
	print(f"{indent}<{node.tag}>")
	if node.text and node.text.strip():
		print(f"{'│   ' * (depth+1)}Text: {node.text.strip()}")
	# Recursive child traversal
	for child in node:
		print_xml_nodes(child, depth + 1)
```

```python
import re
import requests
from bs4 import BeautifulSoup
from typing import Dict, List, Generator

class SEC10QParser:
    def __init__(self):
        self.item_pattern = re.compile(
            # r"item\s*(\d+[A-Za-z]?)\s*[\.\:\-\_]*\s*(.*?)\n",
            r"item\s*(\d+[A-Za-z]?)\xa0",
            re.IGNORECASE | re.DOTALL
        )
        self.section_end_markers = {
            'part': re.compile(r"part\s+(ii|iii|iv)\s*[\.\:\-\_]*", re.IGNORECASE),
            'signature': re.compile(r"signatures?\s*$", re.IGNORECASE)
        }

    def _clean_html(self, html: str) -> str:
        """Extract and clean text from HTML filing"""
        soup = BeautifulSoup(html, 'html.parser')
        for tag in soup(['header', 'footer', 'style', 'script', 'table']):
            tag.decompose()
        return soup.get_text(separator='\n', strip=True)

    def _find_section_bounds(self, text: str) -> Generator[Dict, None, None]:
        """Recursively identify section boundaries using item headers"""
        matches = list(self.item_pattern.finditer(text))
        
        for i, match in enumerate(matches):
            item_num = match.group(1).upper()
            item_title = match.group(2).strip()
            start = match.start()
            
            # Find end of section
            end = None
            next_match = matches[i+1] if i+1 < len(matches) else None
            if next_match:
                end = next_match.start()
            else:
                # Check for document end markers
                part_end = [m.start() for m in self.section_end_markers['part'].finditer(text, start)]
                signature_end = [m.start() for m in self.section_end_markers['signature'].finditer(text, start)]
                end_markers = part_end + signature_end
                end = min(end_markers) if end_markers else len(text)

            yield {
                'item': f"ITEM {item_num}",
                'title': item_title,
                'content': text[start:end].strip(),
                'subsections': text[start:end]
                # 'subsections': list(self._find_section_bounds(text[start:end]))
            }

    def parse_filing(self, url: str) -> List[Dict]:
        """Main parsing method"""
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        
        cleaned_text = self._clean_html(response.text)
        return list(self._find_section_bounds(cleaned_text))
```
